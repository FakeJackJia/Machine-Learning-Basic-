# Decision Tree Regression

image.png

Decision Tree Regression uses the technique of **standard deviation reduction** which means that each split in the dataset would result a smaller standard deviation of the dependent variable corresponding to these features. And each final split is called a **terminal leaf** and an **average** of y-value(dependent variable) is calculated for these final split. And when estimating the values using features, by following the split, a value of y is obtained from the terminal leaf.

This process is pretty similar to a decision tree classifier (check the decision tree classifier for more details) while the only difference is that decision tree regression uses standard deviation reduction as a criteria to check the split. **In short, it would calculate the standard deviation respect to the target variable for the parent node and it will compare with the weighted standard deviation of child nodes after the split; then choose the one with most reduction.** (Note: the way it checks whether it should continue to split or not is uing **coefficient of variation = CV = standard deviation calculated / the average**. Before the algorithm, we will **choose a value for this and any node with a CV less than this criteria will stop splitting**)

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

## Training the Decision Tree Regression model on the whole dataset

```
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

## Predicting a new result

```
regressor.predict([[6.5]])
```

## Visualising the Decision Tree Regression results (higher resolution)

```
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
